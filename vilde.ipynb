```python
import requests

# Din OpenAQ API-nøkkel
api_key = '828ce8950001217e876b9e1e29a137ab72e5bf2a75e5a431f021d67d802eecbf'
city_name = "Berlin"  # Byen du vil hente data fra

# Trinn 1: Finn locations_id for byen
url_locations = f"https://api.openaq.org/v3/locations?city={city_name}"
headers = {"X-API-Key": api_key}

response = requests.get(url_locations, headers=headers)

if response.status_code == 200:
    data = response.json()
    if len(data["results"]) > 0:
        location_id = data["results"][0]["id"]  # Tar første stasjon
        location_name = data["results"][0]["name"]
        print(f"Fant stasjon '{location_name}' med ID {location_id}")

        # Trinn 2: Hent luftkvalitetsdata for den målestasjonen
        url_measurements = f"https://api.openaq.org/v3/locations/{location_id}"
        response_measurements = requests.get(url_measurements, headers=headers)

        if response_measurements.status_code == 200:
            measurements_data = response_measurements.json()
            print(measurements_data)
        else:
            print(f"Feil ved henting av målinger: {response_measurements.status_code}")
    else:
        print(f"Ingen målestasjoner funnet for {city_name}")
else:
    print(f"Feil ved henting av locations: {response.status_code}")
```
